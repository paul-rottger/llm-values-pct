```python
import pandas as pd
```

```python
# load data for relevant experiments and models

EXPERIMENT = "implicit_paraphrase_experiments_230124"
MODELS = ["gpt-3.5-turbo-1106", "Mistral-7B-Instruct-v0.1"]

df_dict = {}

for m in MODELS:
    df_dict[m] = pd.read_csv(f"../data/completions/{EXPERIMENT}/{m}.csv")
```

```python
# sample random entries from each model and combine into one df, then shuffle

N_PER_MODEL = 100

df = pd.concat([df_dict[m].sample(N_PER_MODEL, random_state=42) for m in MODELS])
df = df.sample(frac=1).reset_index(drop=True)

df
```

```python
df.to_csv("../data/annotations/agreement_classification_240124.csv", index=True)
```
